# Q2. Top 5 Car Brands

```python
data = []

for br in df['brand'].unique():
  data.append([br, len(df[df['brand'] == br])])

data = pd.DataFrame(data, columns = ['brand','freq'])
data = data.sort_values(by = 'freq', ascending = False).head()

data.plot(x = 'brand', y = 'freq', kind = 'bar', figsize = (5,5))
```

# Q3. Car Industry Trend

```python
data = []

for yr in df['year'].unique():
  data.append([yr, len(df[df['year'] == yr])])

data = pd.DataFrame(data, columns = ['year','freq'])
data = data.sort_values( by = 'year', ascending = True)

data.plot(x = 'year', y = 'freq', kind = 'line', title = 'Cars Industry Trand')
```

# Q4. Top Car Colours

```python
data = []

for clr in df['color'].unique():
  data.append([clr, len(df[df['color'] == clr])])

data = pd.DataFrame(data, columns = ['color','freq'])
data = data.sort_values( by = 'freq', ascending = False)

data.plot(x = 'color', y = 'freq', kind = 'bar', title = 'Car Colours', figsize = (30,5))
```

# Q5. Which car brad is the most Expensive

```python
plt.figure(figsize = (30,10))
df.groupby('brand').mean()['price'].sort_values(ascending = False).plot.bar()
plt.show()
```

# Q6. Which State is having most expensive cars

```python
plt.figure(figsize = (30,10))
df.groupby('state').mean()['price'].sort_values(ascending = False).plot.bar()
plt.show()
```
